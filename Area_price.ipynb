```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('area_price.csv')
df.head()
```

```python
# we use in matplotlib for visulition 
%matplotlib inline

plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df['area'] , df['price'] , marker=('*') , color='red')
```

```python
# first check the data is clean then we create the model

df.isnull().sum()
```

```python
# create the model
model=linear_model.LinearRegression()
# fitting the model
model.fit(df[['area']],df.price)
```

```python
# predect the value 5000
model.predict([[5000]])
```

```python
model.coef_
```

```python
model.intercept_
```
